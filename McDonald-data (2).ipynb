## 1.Food Category that have the highest and lowest varieties

```python
#Category which has max variety
df.groupby('Category')['Category'].count().idxmax()
```

```python
#Category which has min variety
df.groupby('Category')['Category'].count().idxmin()
```

## 2.Outliers

```python
df.boxplot(figsize=(40,20))
```

## 3.Category that contributes to max percent of Cholestrol in a diet

```python
df.groupby('Category')['Cholesterol (% Daily Value)'].mean().idxmax()
```

## 4.Item which contributes max to sodium intake

```python
df.groupby('Item')['Sodium'].max().idxmax()
```

## 5.Food items which contains most saturated fat

```python
df.groupby('Item')['Saturated Fat'].max().sort_values().tail(4)
```
